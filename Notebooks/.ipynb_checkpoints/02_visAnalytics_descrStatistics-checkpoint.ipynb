### Univariate analysis of claims occurance (binary variable claim weighted against exposure)

```python
# calculate binary variable Claim, which states whether at least 1 claims has happened for the policy
claims_data['Claim'] = claims_data['ClaimNb'].apply(lambda x: 1 if x >=1 else 0)

# Doing a small check
print(claims_data['Claim'].value_counts())
print(claims_data['ClaimNb'].value_counts().sum())
```

```python
# check is ok
678013-34060
```

```python
# calculate the oveall frequency of a claims occurance weighted by esposure
Totfreq_weighted = claims_data['Claim'].sum()/claims_data['Exposure'].sum()
print('overall claim frequency weighted: ', "{0:.2%}".format(Totfreq_weighted))
```

```python
plot_cols = feature_cols[1:]# not use exposure for plotting
plot_cols
```

```python
#  print details for frequency per feature (categorical and numerical with limited nr. of classes)
n_cols = 2 # set number of columns in subplots
n_rows = math.ceil(len(plot_cols) / n_cols) # calculate the respective nr. of rows needed

plt.rc('xtick', labelsize=8) 
plt.rc('ytick', labelsize=8) 

fig, axs = plt.subplots(n_rows, n_cols, figsize = (15,15))
axs2 = axs

for counter in range (len(plot_cols)): # loop over all columns
    col = plot_cols[counter]
    an_table = claims_data.groupby(by = col).agg(n_claims = ('ClaimNb', 'sum'), s_exp = ('Exposure', 'sum')).reset_index()
    an_table['freq_weighted'] = an_table['n_claims'] / an_table['s_exp']
    display(an_table)
    r = counter // n_cols
    c = counter % n_cols
    #print(counter, r,c)
    axs[r,c].bar(an_table[col], an_table['freq_weighted'], label = str(col)) # plot specific requency as bar plot
    axs[r,c].plot(an_table[col], np.ones(len(an_table[col])) * Totfreq_weighted, color = 'red', label= 'average freq')
    axs[r,c].legend()
    
    axs2[r,c] = axs[r,c].twinx()
    axs2[r,c].scatter(an_table[col], an_table['s_exp'], color = 'cyan', marker = '*', label = 'Total Exposure') # plot volume on secondary axis
    axs2[r,c].legend()
    
plt.show()
```

Frequency analysis:
- area: worsening frequency from A-> F -> this has an ordinal meandin, so probably area defines some risk scoring -> take it as ordinal variable to reduce nr. of features in the model
- VehPower: moderate volatility, 4 classes leading from volume point of view
- VehAge: high frequency at age 0 -> volume is low here, but this seems to be an issue here
- Driver Age: young drivers causing more claims, biggest part of portfolio ist between 30 and 60ys
- bonus/Malus: positive correlation visible -> higher malus class -> higher frequency
- Brand: B12: outstandingly worse segment with high volume also
- Density: some regions with outstandingly high frequency but partially driven by low exposure
- Regions: quite volatile
